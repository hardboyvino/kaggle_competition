```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
```

```python
groups = train.groupby('Hardness')

# Sampling 10% from each group
val_indices = groups.apply(lambda x: x.sample(frac=0.1, random_state=5)).index.get_level_values(1)
train_indices = train.index.difference(val_indices)

new_train = train.loc[train_indices].drop('id', axis=1)
# y_train = y.loc[train_indices]
new_val = train.loc[val_indices].drop('id', axis=1)
# y_val = y.loc[val_indices]

new_train.shape, new_val.shape
```
