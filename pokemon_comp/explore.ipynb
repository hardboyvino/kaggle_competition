### generate new abilities features

```python
df_train['train'] = 1
df_test['train'] = 0
```

```python
df_combined = pd.concat([df_train, df_test], axis=0)
```

```python
df_combined.tail()
```

```python
# Use a set to ensure uniqueness
unique_list = []
for item_list in df_combined['abilities']:
    unique_list.append(item_list)

# Initialize empty list for parsed lists
parsed_lists = []

for item in unique_list:
    # Strip brackets and split the string by '
    # Also remove the single quotes around each element
    elements = item.strip("[]").split(", ")
    # Trim the quotes from each element and add to the list
    parsed_lists.append([element.strip("'") for element in elements])

# Flatten the list of lists
flattened_list = [ability for sublist in parsed_lists for ability in sublist]

# Deduplicate to get unique items
unique_abilities = list(set(flattened_list))
unique_abilities
```

```python
for ability in unique_abilities:
    df_combined[f'ability_{ability}'] = 0

df_combined.head()
```

```python
df_combined['abilities'] = df_combined['abilities'].apply(ast.literal_eval)

for index, row in df_combined.iterrows():
    for ability in row['abilities']:
        column_name = f'ability_{ability}'
        if column_name not in df_combined.columns:
            df_combined[column_name] = 0
        df_combined.at[index, column_name] = 1

df_combined.head()
```

```python
# OneHot Encode the type1 features for now
df_combined_onehot_type1 = pd.get_dummies(df_combined['type1'], drop_first=True, prefix='type1').astype(int)
df_combined_onehot_type2 = pd.get_dummies(df_combined['type2'], drop_first=True, prefix='type2').astype(int)
```

```python
# Ensure the capture_rate is in integar
df_combined['capture_rate'] = df_combined['capture_rate'].astype(int)
```

```python
df_combined = pd.concat([df_combined, df_combined_onehot_type1, df_combined_onehot_type2], axis=1)
```

```python
# Get the train and test data back
df_train = df_combined[df_combined['train'] == 1]
df_test = df_combined[df_combined['train'] == 0]

# Drop the train feature from both train and test
df_train.drop(['train', 'abilities', 'japanese_name', 'classfication', 'name', 'pokedex_number', 'type1', 'type2'], axis=1, inplace=True)
df_test.drop(['train', 'abilities', 'japanese_name', 'classfication', 'name', 'pokedex_number', 'type1', 'type2', 'is_legendary'], axis=1, inplace=True)

len(df_train), len(df_test)
```

```python
df_train.to_csv('train.csv', index=False)
df_test.to_csv('test.csv', index=False)
```
