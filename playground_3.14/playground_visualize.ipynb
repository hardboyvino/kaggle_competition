```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the train and test datasets
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
column_to_drop = ['MinOfUpperTRange', 'AverageOfUpperTRange', 'MaxOfLowerTRange', 'MinOfLowerTRange', 'AverageOfLowerTRange','AverageRainingDays']
```

```python
# Remove the columns that are perfectly correlated with each other from train and test

df_train = df_train.drop(column_to_drop, axis=1)
df_test = df_test.drop(column_to_drop, axis=1)
```

```python
df_train.hist(figsize=(20, 15))
plt.show()
```

```python
df_test.hist(figsize=(20, 15))
plt.show()
```

```python
# Pairplot of the entire dataset to observe the correlation between features and also the target variables

sns.pairplot(df_train)
plt.show()
```

```python
corrollation = df_train.corr()
```

```python
plt.figure(figsize=(20, 15))
sns.heatmap(corrollation, xticklabels=corrollation.columns, yticklabels=corrollation.columns, annot=True)
plt.show()
```

```python
# Round the fruitset, fruitmass and see features and then re-visulaize the df_train

df_train = df_train.round({"fruitset": 1, "fruitmass": 1, "seeds": 1, "yield": 1})
```

```python
df_train.hist(figsize=(20, 15))
plt.show()
```

```python
# Check the information distribution for seeds and fruitset after rounding

plt.plot(df_train["seeds"], ".")
plt.show()
```

```python
plt.plot(df_train["fruitset"], ".")
plt.show()
```
